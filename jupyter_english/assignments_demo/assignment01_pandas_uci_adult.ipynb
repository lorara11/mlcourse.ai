**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
print('--- More than 50K ---')
print('Mean of age:', data[data['salary']== '>50K']['age'].mean())
print('Standard deviation of age', data[data['salary']== '>50K']['age'].std())

print('\n--- Less than 50K ---')
print('Mean of age:', data[data['salary']== '<=50K']['age'].mean())
print('Standard deviation of age', data[data['salary']== '<=50K']['age'].std())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
print('--- Married men ---')
married = data[(data.sex == 'Male') & (data['marital-status'].str.startswith('Married'))]['salary'].value_counts()
print(married)
print('Proportion of richer ones:', married[1]/sum(married))
```

```python
print('--- Single men ---')
unmarried = data[(data.sex == 'Male') & (data["marital-status"].isin(["Never-married", "Separated", "Divorced"]))]['salary'].value_counts()
print(unmarried)
print('Proportion of richer ones:', unmarried[1]/sum(unmarried))
```

The proportion of those who earn more than 50K is greater among married men. Though it wasn't asked, let's see if the same holds for women (or females).

```python
print('--- Married women ---')
marriedw = data[(data.sex == 'Female') & (data['marital-status'].str.startswith('Married'))]['salary'].value_counts()
print(marriedw)
print('Proportion of richer ones:', marriedw[1]/sum(marriedw))
```

```python
print('--- Single women ---')
unmarriedw = data[(data.sex == 'Female') & (data["marital-status"].isin(["Never-married", "Separated", "Divorced"]))]['salary'].value_counts()
print(unmarriedw)
print('Proportion of richer ones:', unmarriedw[1]/sum(unmarriedw))
```

Yes, it does hold for women/females as well.

I also see that, in this dataset, the proportion of single women compared to married women is a lot greater than the ratio of single men compared to married men. That was surprising.

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_hours_week = data['hours-per-week'].max()
print('Max number of hours per week:', max_hours_week)
num_workers = data[data['hours-per-week'] == max_hours_week].shape[0]
print(num_workers, 'people work',max_hours_week, 'hours per week.')

ratio_richer = data[(data['hours-per-week'] == max_hours_week) & (data.salary == '>50K')].shape[0] / num_workers
print('The proportion of those who earn >50K among them is', ratio_richer)
```
